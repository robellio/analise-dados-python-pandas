##**Analise Exploratória com python e pandas**

```python
#Fazendo importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Criando DataFrame
df = pd.read_excel("/content/drive/MyDrive/Datasets/AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Verificando a quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a Receita Total?
df["Valor Venda"].sum()
```

```python
#Qual o Custo Total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
#Qual o Custo total?
round(df["custo"].sum(), 2)
```

```python
#Qual o Lucro Total?
df["lucro"] = df["Valor Venda"] -df["custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
round(df["lucro"].sum(),2)
```

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

**Transformando a coluna Tempo_envio em númerica e Calculando a média do tempo de envio para cada marca.**

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

#**Missing Values**

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

#**Lucro por Ano e por Marca**

```python
#Agrupando por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format='{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

##**Gráfico Total de Produtos Vendidos**

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

##**Gráfico de Lucro por Ano**

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro X Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

##**Gráfico de Lucro por mês**

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

##**Gráfico de Lucro por Marca**

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

##**Gráfico de Lucro por Classe**

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
